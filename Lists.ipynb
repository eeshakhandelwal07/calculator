List of roll numbers

```python
L1 = [101,102,103,104,105]
print("The roll nos. are" , L1)
print(type(L1))
```

To add new items to list (append = up in the end)

```python
#TO ADD NEW ITEMS TO LIST
L1.append(110)
print(L1)
```

To change at a particular location


```python
#we cannot ue append as it is not in the end
#here we use insert

L1.insert(2,106)
print(L1)
```

To delete items

delete in the end del specifically

```python
#to remove last element use pop
L1.pop()
print(L1)
```

To delete from specific location

```python
#here we use remove
L1.remove(106)
print(L1)
```
